```python
import pandas as pd
import numpy as np

import plotly.express as px
import plotly.graph_objects as go
```

```python
df = px.data.gapminder()
df
```

```python
fig = go.FigureWidget(
    data={
        'type': 'scattergl',
        'x': df['pop'],
        'y': df['gdpPercap'],
        'mode': 'markers',
    }
)
fig
```

```python
fig.layout.title = 'gdppc vs pop'
fig.layout.titlefont.size = 20
fig.layout.titlefont.family = 'Rockwell'

fig.layout.yaxis.title = 'gdppc'
fig.layout.xaxis.title = 'pop'

scatter = fig.data[0]
```

```python
from ipywidgets import interactive

def set_opacity(opacity, size):
    scatter.marker.opacity = opacity
    scatter.marker.size = size

opacity_slider = interactive(set_opacity, opacity=(0, 1, 0.1), size=(1, 10, 0.25))
opacity_slider
```

```python
from ipywidgets import HTML

details = HTML()

def hover_fn(trace, points, state):
    ind = points.point_inds[0]
    details.value = df.iloc[ind].to_frame().to_html()

scatter.on_hover(hover_fn)
details
```

```python
from ipywidgets import VBox, HBox

HBox([fig, opacity_slider, details])
```

```python
fig = go.FigureWidget(
    data={
        'type': 'scattergl',
        'x': df['pop'],
        'y': df['gdpPercap'],
        'mode': 'markers',
    }
)
scatter = fig.data[0]

hfig = go.FigureWidget().add_histogram(x=df['gdpPercap'])
hist = hfig.data[0]

def update_hist(trace, points, state):
    if points.point_inds:
        hist.x = df.gdpPercap.iloc[points.point_inds]
    else:
        hist.x = df.gdpPercap

scatter.on_selection(update_hist)

display(fig, hfig)
```

```python
s = px.scatter(
    df,
    x='pop', y='lifeExp',
    color='year',
    symbol='continent',
    # size='pop',
    hover_name='country',
    hover_data='iso_alpha',
    # text='country',

    title='testing',
    color_continuous_scale='magma',
    template='simple_white'
)

fig = go.FigureWidget(s)

def update_title(trace, points, selector):
    # display(trace, points, selector)
    if points.point_inds:
        index = points.point_inds[0]
        label = df.loc[index, 'country']
        fig.update_layout(title=f'hovered over: {label}')
    else:
        fig.update_layout(title='hover...!')
        # fig.layout.title.text = 'hover///'


fig.data[0].on_hover(update_title)
fig
```

```python
import plotly.graph_objects as go

import numpy as np
np.random.seed(1)

x = np.random.rand(100)
y = np.random.rand(100)

f = go.FigureWidget([go.Scatter(x=x, y=y, mode='markers')])

scatter = f.data[0]
colors = ['#a3a7e4'] * 100
scatter.marker.color = colors
scatter.marker.size = [10] * 100
f.layout.hovermode = 'closest'


# create our callback function
def update_point(trace, points, selector):
    c = list(scatter.marker.color)
    s = list(scatter.marker.size)
    for i in points.point_inds:
        c[i] = '#bae2be'
        s[i] = 20
        print(i)
        with f.batch_update():
            scatter.marker.color = c
            scatter.marker.size = s


scatter.on_hover(update_point)

f
```
